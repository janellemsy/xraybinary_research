## This notebook is used to plot data produced by forward simulating the suspected progenitor of NGC 300 X-1 using MESA (Modules for Experiments in Stellar) Astrophysics). It is meant to track certain characteristics of the system as a function of time. We want to figure out what type of progenitor system could create the current system NGC 300 X-1 has.

```python
%matplotlib inline
import numpy as py
from numpy import *
import matplotlib.pyplot as plt

#load & plot data
data = loadtxt('100_45_16.5_0.007_binary.dat')
x1 = data[:,0]
y1 = data[:,3]
y1_2 = data[:,4]
plt.figure()
plt.plot(x1, y1, color = 'k', label = 'Star 1: $M_i$ = 100.0')
plt.plot(x1, y1_2, color = 'b', label = 'Star 2: $M_i$ = 45.0')
plt.xlim(0)
plt.ylim(0)
plt.title('Mass Versus Time for NGC 300 X-1')
plt.xlabel('Time (Myr)')
plt.ylabel(r'Mass ($M_\odot$)')
plt.legend()
plt.grid()

#load & plot data  
data = loadtxt('95_45_16.35_0.007_binary.dat')
x2 = data[:,0]
y2_1 = data[:,3]
y2_2 = data[:,4]
plt.figure()
plt.plot(x2, y2_1, color = 'k', label = 'Star 1: $M_i$ = 95.0')
plt.plot(x2, y2_2, color = 'b', label = 'Star 2: $M_i$ = 45.0')
plt.xlim(0)
plt.ylim(0)
plt.title('Mass Versus Time for NGC 300 X-1')
plt.xlabel('Time (Myr)')
plt.ylabel(r'Mass ($M_\odot$)')
plt.legend()
plt.grid()

#load & plot data  
data = loadtxt('90_45_16.15_0.007_binary.dat')
x3 = data[:,0]
y3_1 = data[:,3]
y3_2 = data[:,4]
plt.figure()
plt.plot(x3, y3_1, color = 'k', label = 'Star 1: $M_i$ = 90.0')
plt.plot(x3, y3_2, color = 'b', label = 'Star 2: $M_i$ = 45.0')
plt.xlim(0)
plt.ylim(0)
plt.title('Mass Versus Time for NGC 300 X-1')
plt.xlabel('Time (Myr)')
plt.ylabel(r'Mass ($M_\odot$)')
plt.legend()
plt.grid()

plt.show()
```
