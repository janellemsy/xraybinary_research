```python
#x1dsum plot using text file

#First April 27 file
data1 = loadtxt('le6g03010_x1dsum_April27.txt')
x1 = data1[:,0]
y1 = data1[:,1]

#Second April 27 file
data2 = loadtxt('le6g04010_x1dsum_April27.txt')
x2 = data2[:,0]
y2 = data2[:,1]

#plotting both datasets
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(x1, y1,color='k')

#making 1st plot pretty
plt.title('1st Dataset for April 27')
plt.xlim(1100,2550)
plt.ylim(-1.0e-14,3e-14)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$)')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x2, y2,color='k')
#making 2nd plot pretty
plt.title('2nd Dataset for April 27')
plt.xlim(1100,2550)
plt.ylim(-1.0e-14,3e-14)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$)')
plt.grid()
```

```python
#GAUSSIAN FITTING
#geochromal airglow line (ignore for now)~

# plotting data
data = 'le6g03010_x1dsum_April27.txt'
sp = pyspeckit.Spectrum(data)

#limits and axes
sp.plotter(xmin = 1290, xmax = 1320,ymax = 1e-14, color = 'k')
plt.title('April 27 Gaussian #1 Fit')
sp.plotter.axis.set_xlabel(r'Wavelength $(\AA)$')
sp.plotter.axis.set_ylabel(r'Flux ($\mathrm{erg/s/cm^2/\AA})$')
plt.grid()
sp.plotter.refresh()

#Fitting with automatic guesses
#sp.specfit(fittype='gaussian', annotate = False)

#fitting with my own guesses
sp.specfit(fittype='gaussian',
           guesses = [6.1e-15,1306,5,
                      4.0e-15,1304,2,
                      3.8e-15,1302,2,
                      2e-15, 1300,3], annotate = False)


print(sp.specfit.parinfo)
```

```python
import matplotlib.pyplot as plt
import pyspeckit 
import numpy as np

#GAUSSIAN FITTING

# plotting data
data = 'le6g03010_x1dsum_April27.txt'
sp = pyspeckit.Spectrum(data)

#limits and axes
sp.plotter(xmin = 1620, xmax = 1660,ymax = 2.5e-15, color = 'k')
plt.title('April 27 Gaussian #2 Fit')
sp.plotter.axis.set_xlabel(r'Wavelength $(\AA)$')
sp.plotter.axis.set_ylabel(r'Flux ($\mathrm{erg/s/cm^2/\AA})$')
plt.grid()
sp.plotter.refresh()

#Fitting with automatic guesses
#sp.specfit(fittype='gaussian', annotate = False)

#fitting with my own guesses
sp.specfit(fittype='gaussian',
           guesses = [1.0e-15,1642,10], annotate = True)

sp.specfit.plot_components(xarr=None, show_hyperfine_components=None, component_yoffset=0.08e-15)
print(sp.specfit.parinfo)
```
