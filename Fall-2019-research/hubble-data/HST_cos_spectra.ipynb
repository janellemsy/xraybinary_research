```python
from astropy.io import fits                                      
import numpy as np
```

```python
#converting fits file to txt file
x1ddata = fits.getdata('le6g04010_x1dsum_April27.fits')
wavelength = np.concatenate((x1ddata["wavelength"][1],\
              x1ddata["wavelength"][0]))
flux = np.concatenate((x1ddata["flux"][1], x1ddata["flux"][0]))
err = np.concatenate((x1ddata["error"][1], x1ddata["error"][0]))
dq_wgt = np.concatenate((x1ddata["dq_wgt"][1], \
         x1ddata["dq_wgt"][0]))
with open('le6g04010_x1dsum_April27.txt','w') as f:
    for row in zip(wavelength, flux, err, dq_wgt):
        f.write('{}\t{}\t{}\t{}\n'.format(row[0], row[1], row[2], row[3]))
```

```python
import matplotlib.pyplot as plt
from astropy.io import fits                                      
import numpy as np
from numpy import *

#x1dsum plot using text file

#April 27
data1 = loadtxt('le6g03010_x1dsum_April27.txt')
x1 = data1[:,0]
y1 = data1[:,1]

#2nd data set for April 27
data2 = loadtxt('le6g04010_x1dsum_April27.txt')
x2 = data2[:,0]
y2 = data2[:,1]

#April 29
data3 = loadtxt('le6g02010_x1dsum_April29.txt')
x3 = data3[:,0]
y3 = data3[:,1]

#May 1
data4 = loadtxt('le6g01010_x1dsum_May1.txt')
x4 = data4[:,0]
y4 = data4[:,1]

#plotting April 27
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(x1, y1,color='k')
plt.xlim(1100,2550)
plt.ylim(-1.0e-14,3e-14)
plt.grid()

#axes & title
plt.title('1st Dataset for April 27', y=1.05)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$)')

#plotting 2nd dataset for April 27
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(x2, y2,color='r')
plt.xlim(1100,2550)
plt.ylim(-1.0e-14,3e-14)
plt.grid()

#axes & title
plt.title('2nd Dataset for April 27', y=1.05)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$)')

#plotting April 29
plt.figure(figsize=(12,8))
plt.subplot(2,2,2)
plt.plot(x3, y3,color='b')
plt.xlim(1100,2550)
plt.ylim(-1.0e-14,3e-14)
plt.grid()

#axes & title
plt.title('April 29', y=1.05)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$)')

#plotting May 1
plt.figure(figsize=(12,8))
plt.subplot(2,2,3)
plt.plot(x4, y4,color='g')
plt.xlim(1100,2550)
plt.ylim(-1.0e-14,3e-14)

#axes & title
plt.title('May 1', y=1.05)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$)')
plt.grid()


plt.show()
```

```python
import matplotlib.pyplot as plt
import pyspeckit 
import numpy as np

#GAUSSIAN FITTING

# plotting data
data = 'le6g03010_x1dsum_April27.txt'
sp = pyspeckit.Spectrum(data)

#limits and axes
sp.plotter(xmin = 1290, xmax = 1320,ymax = 1e-14, color = 'k')
plt.title('April 27 Gaussian Fit')
sp.plotter.axis.set_xlabel(r'Wavelength $(\AA)$')
sp.plotter.axis.set_ylabel(r'Flux ($\mathrm{erg/s/cm^2/\AA})$')
plt.grid()
sp.plotter.refresh()

#Fitting with automatic guesses
#sp.specfit(fittype='gaussian', annotate = False)

#fitting with my own guesses
sp.specfit(fittype='gaussian',
           guesses = [6.1e-15,1306,5,
                      4.0e-15,1304,2,
                      3.8e-15,1302,2,
                      2e-15, 1300,3], annotate = False)


print(sp.specfit.parinfo)
```
