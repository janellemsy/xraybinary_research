```python
import matplotlib.pyplot as plt
from astropy.io import fits                                      
import numpy as np
from numpy import *
import pyspeckit
```

```python
#rebinning data

data = loadtxt('le6g01010_x1dsum_May1.txt')
x = data[:,0]
y = data[:,1]

from scipy.ndimage.filters import gaussian_filter1d
spectrum = y
smoothed_spectrum = gaussian_filter1d(spectrum, sigma=3)
```

```python
#testing out plotting rebinned data

plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
plt.title('May 1 Spectrum')
plt.plot(x, smoothed_spectrum,color='k')
plt.xlim(1100,2550)
plt.ylim(-1.0e-14,1.5e-14)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$)')
plt.grid()
plt.show()
```
