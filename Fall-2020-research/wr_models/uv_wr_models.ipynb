```python
%matplotlib inline
import numpy as py
from numpy import *
import matplotlib.pyplot as plt


#loading in data for WNE Model 07-18
data2 = loadtxt('lmc-wne_07-18_sed.txt')
x2 = data2[:,0]
y2 = data2[:,1]
flux2 =(10**y2)
F2 = (2.5e-11*flux2)
lambda2 = (10**x2)

#loading in data for WNE Model 09-04
data1 = loadtxt('lmc-wne_09-04_sed.txt')
x1 = data1[:,0]
y1 = data1[:,1]
flux=(10**y1)
F1 = (2.5e-11*flux)
lambda1 = (10**x1)

#loading in data for WNE Model 11-04
data3 = loadtxt('lmc-wne_11-04.txt')
x3 = data3[:,0]
y3 = data3[:,1]
flux3=(10**y3)
F3 = (2.5e-11*flux3)
lambda3 = (10**x3)

#loading in data for WNE Model 12-15
data4 = loadtxt('lmc-wne_12-15_sed.txt')
x4 = data4[:,0]
y4 = data4[:,1]
flux4 =(10**y4)
F4 = (2.5e-11*flux4)
lambda4 = (10**x4)

#plotting 07-18
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(lambda2, F2,color='r')
plt.xlim(900,2391)

#axes & title
plt.title('UV Spectrum of WNE Model 07-18 (T=50.1 kK)',y=1.05)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$) in d=2Mpc ')

#plotting 09-04
plt.subplot(2,2,2)
plt.plot(lambda1, F1, color='r')
plt.xlim(900,2391)

#axes & title
plt.title('UV Spectrum of WNE Model 09-04 (T=63.1 kK)',y=1.05)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$) in d=2Mpc ')

#plotting 11-04
plt.subplot(2,2,3)
plt.plot(lambda3, F3,color='r')
plt.xlim(900,2391)

#axes & title
plt.title('UV Spectrum of WNE Model 11-04 (T=79.4 kK)',y=1.05)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$) in d=2Mpc ')

#plotting 12-15
plt.subplot(2,2,4)
plt.plot(lambda4, F4,color='r')
plt.xlim(900,2391)

#axes & title
plt.title('UV Spectrum of WNE Model 12-15 (T=89.1 kK)',y=1.05)
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$) in d=2Mpc ')


plt.tight_layout()
plt.show()
```

```python
#plotting the UV spectrum of all the models at d=2Mpc
plt.figure(figsize=(10,6))

#07-18
plt.plot(lambda2,F2, color='c',label='WNE 07-18 (T=50.1 kK)')
plt.legend()

#09-04
plt.plot(lambda1,F1, color='r',label='WNE 09-04 (T=63.1 kK)')
plt.legend()

#11-04
plt.plot(lambda3,F3, color='m',label='WNE 11-04 (T=79.4 kK)')
plt.legend()

#12-15
plt.plot(lambda4,F4, color='g',label='WNE 12-15 (T=89.1 kK)')
plt.legend()

#title and axes labels
plt.title('UV Spectrum of all WNE Models at d=2Mpc')
plt.xlabel('Wavelength \u03BB (\u212B)')
plt.ylabel('$F_\u03BB$ (erg s$^{-1}$ cm$^{-2}$ \u212B$^{-1}$) in d=2Mpc ')

#setting axes limits
plt.xlim(900,2391)
plt.ylim(0,0.5e-14)

plt.show()
```
